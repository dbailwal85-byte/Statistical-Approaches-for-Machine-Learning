```python

import numpy as np

# Grouped data: Age groups and frequencies
age_groups = [(15,25), (25,35), (35,45), (45,55), (55,65)]
frequency = np.array([350, 375, 120, 80, 90], dtype=float)

# Class midpoints
midpoints = np.array([(a + b)/2 for a, b in age_groups])

# Class width (h)
h = age_groups[0][1] - age_groups[0][0]


# Mean Calculation
N = frequency.sum()
mean = np.sum(midpoints * frequency) / N

# Median Calculation
# Cumulative frequency
cum_freq = np.cumsum(frequency)
N_by_2 = N / 2

# Find median class
median_class_index = np.searchsorted(cum_freq, N_by_2)
L = age_groups[median_class_index][0] - 0.5  # lower boundary
C = cum_freq[median_class_index - 1] if median_class_index > 0 else 0
f_m = frequency[median_class_index]

median = L + ((N_by_2 - C) / f_m) * h

# Mode Calculation
modal_index = int(np.argmax(frequency))
f1 = frequency[modal_index]
f0 = frequency[modal_index - 1] if modal_index > 0 else 0
f2 = frequency[modal_index + 1] if modal_index < len(frequency) - 1 else 0
L_mode = age_groups[modal_index][0] - 0.5

mode = L_mode + ((f1 - f0) / (2 * f1 - f0 - f2)) * h

# -------------------------------
# Final Output
# ------------------------------
print("------ STATISTICAL RESULTS ------")
print(f"Total Frequency (N): {int(N)}")
print(f"Mean   = {mean:.2f}")
print(f"Median = {median:.2f}")
print(f"Mode   = {mode:.2f}")
```
